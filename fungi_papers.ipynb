```python
import pandas as pd

with open('available_genomes.txt', 'r') as f:
    lines = f.readlines()
    species = [line.split("\t")[1] for line in lines[1:]]
    accessions = [line.split("\t")[0] for line in lines[1:]]
species_df = pd.DataFrame({"species": species, "accession": accessions})
species_df.to_csv("fungi_species.csv", index=False)
```

```python
from types import NoneType
import requests
from bs4 import BeautifulSoup #html parser
res = requests.get("https://mycocosm.jgi.doe.gov/mycocosm/home/releases?flt=aspergillus%20ruber").text #page in mycocasm

html = BeautifulSoup(res, "html.parser")
trs = html.find_all("tr") #finds all td tags which will give list of all nodes, this one for rows in the table (tr=tablerow)
for tr in trs:
    tds = tr.find_all("td") #finds all td tags
    #print(tds)
    link = tds[5].find("a") if len(tds) > 4 else None
    link = link.attrs["href"] if link else None
    print(link)
```
